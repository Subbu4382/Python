Output: [4, 5, 1, 2, 3]

```python
arr=[1,2,3,4,5]
k=int(input())
def rotate_rg(arr,k):
    l=arr[-k:] + arr[0:-k]
    return l
rotate_rg(arr,k)
```

Intersection of Two Arrays

Problem: Write a function that returns the common elements between two arrays.
Testcase 1:
Input: [1, 2, 3], [2, 3, 4]
Output: [2, 3]

```python
def common_ele(arr1,arr2):
    common=[]
    small=arr2 if len(arr1)>=len(arr2) else arr1
    big=arr1 if len(arr1)>=len(arr2) else arr2
    for i in small :
        if i in big:
            common.append(i)
    return common
common_ele([1,2,3],[2,3,4])
```

Find Missing Number

Problem: Given an array of consecutive numbers with one missing, find the missing number.
Testcase 1:
Input: [1, 2, 4, 5]
Output: [3]

```python
def missing_num(arr):
    for i in range(len(arr)-1):
        if arr[i]+1!=arr[i+1]:
            return arr[i]+1
missing_num([1,2,4,5])
```

Find the Maximum Product of Two Elements

Problem: Write a function to find the maximum product of two elements in an array.
Testcase 1:
Input: [3, 5, -2, 8, 11]
Output: 8 * 11 → 88


```python
arr=[3,5,-2,8,11]
max=0
for i in range(len(arr)):
    for j in range(i+1,len(arr)):
        product=arr[i]*arr[j]
        if product>max:
            max=product
print(max)
```

Move Zeros to End

Problem: Write a function that moves all zeros in an array to the end while maintaining the order of other elements.
Testcase 1:
Input: [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]


```python
def zeros(arr):
    non_zero=[]
    zeros=[]
    for i in arr:
        if i!=0:
            non_zero.append(i)
        else:
            zeros.append(i)
    res=non_zero+zeros
    return res
zeros([0,1,0,3,12])
```

Pair Sum

Problem: Write a function to find all pairs in an array whose sum is equal to a given target.
Testcase 1:
Input: [2, 4, 3, 5, 7, 8, 9], 7
Output: [[4, 3], [2, 5]]

```python
def pairs(arr,target):
    res=[]
    print(arr,target)
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i]+arr[j]==target:
              res.append([arr[i],arr[j]])
    return res
pairs([2,4,3,5,7,8,9],7)
```

Find Peak Element

Problem: Write a function to find a peak element in an array. An element is a peak if it is not smaller than its neighbours.
Testcase 1:
Input: [1, 3, 20, 4, 1, 0]
Output: 20


```python
arr=[1,3,20,4,1,0]
def peak(arr):
    for i in range(len(arr)):
        if i==0 and arr[i]>arr[i+1]:
            return arr[i]
        elif i==len(arr) and arr[i]>arr[i-1]:
            return arr[i]
        elif arr[i]>arr[i-1] and arr[i]>arr[i+1]:
            return arr[i]
peak(arr)

```

Find the First Duplicate

Problem: Write a function to return the first duplicate value in an array.
Testcase 1:
Input: [2, 1, 3, 5, 3, 2]
Output: 3

```python
def first_duplicate(arr):
    res=[]
    duplicate=[]
    for i in arr:
        if i not in res:
            res.append(i)
        else:
            if i not in duplicate:
                duplicate.append(i)
    return f"first duplicate value is {duplicate[0]}"
first_duplicate([2,1,3,5,3,2])
```

Group Anagrams

Problem: Write a function to group anagrams from an array of strings.
Testcase 1:
Input: ["eat", "tea", "tan", "ate", "nat", "bat"]
Output: [["eat", "tea", "ate"], ["tan", "nat"], ["bat"]]


```python
def groupAnagrams(words):
    groups = {}

    for w in words:
        freq = [0] * 26
        
        for ch in w:
            freq[ord(ch) - ord('a')] = 1
        
        key = str(freq)

        if key not in groups:
            groups[key] = []
        
        groups[key].append(w)

    return list(groups.values())
    
print(groupAnagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
```

Longest Increasing subsequence in an array.
Testcase 1:
Input: [10, 9, 2, 5, 3, 7, 101, 18]

```python
arr = [10, 9, 2, 5, 3, 7, 101, 18]

dp = [1] * len(arr)   # every element has LIS = 1
longest = 1           # store final result manually

for i in range(1, len(arr)):
    for j in range(i):
        if arr[j] < arr[i]:       # increasing condition
            # if dp[j] + 1 > dp[i]: # manual max
                dp[i] = dp[j] + 1

    # update longest manually (instead of using max())
    if dp[i] > longest:
        longest = dp[i]
print(dp[-1])

print(longest)
```
